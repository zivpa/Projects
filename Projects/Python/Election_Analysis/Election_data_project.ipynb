```
# Get votes of all parties (normalized)
def parties_votes(df, thresh):
    par = round(df.sum().div(df.sum().sum()).sort_values(ascending=False),5)
    return par[par > thresh]

part_percent_2020 = parties_votes(ballot_2020_parties, 0.0325)
part_percent_2019b = parties_votes(ballot_2019b_parties, 0.0325)
part_percent_2019a = parties_votes(ballot_2019a_parties, 0.0325)

shas_percent = pd.DataFrame([part_percent_2019a['שס'],part_percent_2019b['שס'],part_percent_2020['שס']])
shas_percent.index = ['April 2019', "September 2019", "2020"]
shas_percent.plot(kind='bar')
plt.title("Vote percent סש got at each election")
plt.xlabel('מועד בחירות'[::-1])
plt.ylabel('אחוז הקולות ש-"שס" קיבלו'[::-1])
print("The vote percentage for 'שס' in each election, was:  in April 2019", part_percent_2019a['שס'], "in September 2019", part_percent_2019b['שס'], "in 2020", part_percent_2020['שס'])
```

```
ballot_2019b['שם ישוב'] = ballot_2019b.index
ballot_2019a_2019b = pd.merge(left=ballot_2019a, right=ballot_2019b, how='inner', left_on=['סמל ישוב','מספר קלפי'], right_on=['סמל ישוב','קלפי'], suffixes=(' 2019a', ' 2019b'))
ballot_2019a_2019b.index = ballot_2019a_2019b['שם ישוב']

ballot_2019a_clean = ballot_2019a_2019b.iloc[:,2:13]
ballot_2019b_clean = ballot_2019a_2019b.iloc[:,16:26]


ballot_2019a_clean.columns = ballot_2019a.columns[2:13]
ballot_2019b_clean.columns = ballot_2019b.columns[2:12]

ballot_2019a_clean['לא הצביעו'] = ballot_2019a_2019b['בזב 2019a'] - ballot_2019a_2019b['כשרים 2019a']
ballot_2019b_clean['לא הצביעו'] = ballot_2019a_2019b['בזב 2019b'] - ballot_2019a_2019b['כשרים 2019b']


ballot_2019a_rev_names = [x[::-1] for x in ballot_2019a_clean.columns]
ballot_2019b_rev_names = [x[::-1] for x in ballot_2019b_clean.columns]

M_2019a_2019b = pd.DataFrame(np.dot(np.dot(np.linalg.pinv(np.dot(ballot_2019a_clean.T, ballot_2019a_clean)), ballot_2019a_clean.T), ballot_2019b_clean))
M_2019a_2019b[M_2019a_2019b < 0.005] = 0
M_2019a_2019b_norm = round(M_2019a_2019b.div(M_2019a_2019b.sum(axis=1), axis=0), 3)
M_2019a_2019b_norm.index = ballot_2019a_rev_names
M_2019a_2019b_norm.columns = ballot_2019b_rev_names

fig, ax = plt.subplots(figsize = (12,12))
plt.rcParams.update({'font.size': 14})
plot = sns.heatmap(M_2019a_2019b_norm, linewidths=0.10, annot=True)
plt.ylabel("april 2019")
plt.xlabel("september 2019")
plt.title("parties vote transformation between election")
sm.OLS(ballot_2019b_clean['שס'],ballot_2019a_clean).fit().params.to_dict()
summary = sm.OLS(ballot_2019b_clean['שס'],ballot_2019a_clean)
print(summary.fit().summary())
```
